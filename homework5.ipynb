```python
#W5A3
def primenumber(a):
    if a<2:
        return False    
    for i in range(2,int(a**0.5)+1):
        if a%i==0:
            return False
        else:
            return True
    
```

```python
#W5A4
def perfectnumber(n):
    tong=0
    for i in range(1,n):
        if n%i==0:
            tong+=i
    if tong==n:
        return True
    else:
        return False
perfectnumber(28)
```

```python
#W5A5
def songuyen(n,k):
    for i in str(n):
        if int(i)==k:
            vt=str(n).index(str(k))
        return vt+1
    return -1
songuyen(123456,5)
```

```python
#
def may_tinh_bo_tui(num1, num2, operat):

    ket_qua = 0
    
    if operat == '+':
        ket_qua = num1 + num2
    elif operat == '-':
        ket_qua = num1 - num2
    elif operat == '*':
        ket_qua = num1 * num2
    elif operat == '/':
        if num2 == 0:
            return "Lỗi: Không thể chia cho 0"
        ket_qua = num1 / num2
    else:
        return "Lỗi: Toán tử không hợp lệ"
    return round(ket_qua, 2)

# --- Phần chạy thử (Test) ---
n1 = float(input("Nhập số thứ nhất: "))
n2 = float(input("Nhập số thứ hai: "))
op = input("Nhập phép toán (+, -, *, /): ")

result = may_tinh_bo_tui(n1, n2, op)
print(f"Kết quả: {result}")
```
