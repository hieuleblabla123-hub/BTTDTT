```python
import math

def kiem_tra_nguyen_to(n):
    """Hàm kiểm tra một số có phải là số nguyên tố hay không"""
    # Số nguyên tố phải lớn hơn 1
    if n < 2:
        return False
    
    # Chỉ cần kiểm tra từ 2 đến căn bậc 2 của n
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False # Nếu chia hết cho số nào đó thì không phải nguyên tố
    return True

def main():
    # Phần 1: Nhập cho đến khi được số nguyên dương
    while True:
        try:
            n = int(input("Nhập một số nguyên dương: "))
            if n > 0:
                break # Thoát vòng lặp nếu nhập đúng
            else:
                print("Vui lòng nhập số lớn hơn 0!")
        except ValueError:
            print("Đó không phải là số nguyên. Vui lòng nhập lại!")

    # Phần 2: Kiểm tra và in kết quả
    if kiem_tra_nguyen_to(n):
        print(f"{n} là số nguyên tố.")
    else:
        print(f"{n} không phải là số nguyên tố.")

if __name__ == "__main__":
    main()
```
