```python
# bai 8 
hang1=(input(" nhap a,b,m")).split(" ")
a,b,m= map(float,hang1)

hang2=(input(" nhap c,d,n ")).split(" ")
c,d,n= map(float,hang2)
# --- 2. Tính các Định thức ---
D = a * d - b * c
Dx = m * d - b * n
Dy = a * n - m * c

print(f"Hệ số: a={a}, b={b}, m={m}, c={c}, d={d}, n={n}")

if D != 0:
    x = Dx / D
    y = Dy / D
    
    print("--- Nghiệm của hệ phương trình (Cramer) ---")
    print(f"x = {x}")
    print(f"y = {y}")

elif D == 0 and (Dx != 0 or Dy != 0):
    # D = 0 nhưng Dx hoặc Dy khác 0
    print("Hệ phương trình VÔ NGHIỆM.")
    
elif D == 0 and Dx == 0 and Dy == 0:
    # D = Dx = Dy = 0
    print("Hệ phương trình VÔ SỐ NGHIỆM.")
```

```python

print("--- Bài 10: Kiểm tra điểm thuộc đường tròn ---")
# Nhập tọa độ tâm I và bán kính R
xi = float(input("Nhập tọa độ tâm Ix: "))
yi = float(input("Nhập tọa độ tâm Iy: "))
r = float(input("Nhập bán kính R: "))

# Nhập tọa độ điểm A
xa = float(input("Nhập tọa độ điểm Ax: "))
ya = float(input("Nhập tọa độ điểm Ay: "))

# --- 2. Tính toán và Kiểm tra ---
# Tính bình phương khoảng cách d^2
distance_sq = (xa - xi)**2 + (ya - yi)**2

# Tính bình phương bán kính R^2
radius_sq = r**2

# So sánh bình phương khoảng cách với bình phương bán kính.
# Dùng math.isclose để so sánh float, tránh lỗi sai số nhỏ
import math
epsilon = 1e-6 

if math.isclose(distance_sq, radius_sq, abs_tol=epsilon):
    print(f"Điểm A({xa}, {ya}) THUỘC đường tròn (cách tâm I một khoảng bằng R).")
elif distance_sq < radius_sq:
    print(f"Điểm A({xa}, {ya}) nằm BÊN TRONG đường tròn.")
else:
    print(f"Điểm A({xa}, {ya}) nằm BÊN NGOÀI đường tròn.")
```
