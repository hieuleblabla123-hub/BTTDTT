```python
#W3A11
a = int(input("Nhập độ dài cạnh a: "))
b = int(input("Nhập độ dài cạnh b: "))
c = int(input("Nhập độ dài cạnh c: "))

# BƯỚC 1: KIỂM TRA ĐIỀU KIỆN TAM GIÁC
# Điều kiện: Tổng hai cạnh bất kỳ phải lớn hơn cạnh còn lại
if (a + b > c) and (a + c > b) and (b + c > a):
    # Nếu là tam giác hợp lệ, chuyển sang BƯỚC 2: PHÂN LOẠI
    
    # 2.1: Kiểm tra Tam giác đều (ba cạnh bằng nhau)
    if (a == b) and (b == c):
        print("Tam giác đều")
        
    # 2.2: Kiểm tra Tam giác cân (hai cạnh bằng nhau)
    # Dùng 'elif' vì nếu đã là Tam giác đều thì không cần kiểm tra nữa
    elif (a == b) or (b == c) or (a == c):
        print("Tam giác cân")
        
    # 2.3: Trường hợp còn lại là Tam giác thường (ba cạnh khác nhau)
    else:
        print("Tam giác thường")
        
else:
    # Nếu không thỏa mãn điều kiện tam giác
    print("Không phải tam giác")
```
