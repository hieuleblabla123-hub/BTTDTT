```python
from jupyturtle import forward, left, right, back

def draw(length):
    angle = 50
    factor = 0.6

    if length > 5:
        forward(length)
        left(angle)
        draw(factor * length)
        right(2 * angle)
        draw(factor * length)
        left(angle)
        back(length)
```

```python
def checkperfect_number(n)-> bool:
    if n < 1:
        return False
    divisors = [i for i in range(1, n) if n % i == 0]
    print(bool(divisors))
    return sum(divisors) == n
print(checkperfect_number(28))  
```

```python
def linhtinh(a):
    tong=0
    so=["1","2","3","4","5","6","7","8","9","0"]
    cacso=[]
    for i in str(a):
        if str(i) in so:
            cacso.append(i)
    for j in cacso:
        tong+= int(j)
    return tong
linhtinh("aaaaaaa")
if linhtinh("aaaaaaa") ==0:
    print ("-1")
```
