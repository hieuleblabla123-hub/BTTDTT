```python
#bài 5
# 1. Nhập liệu
# Bỏ str() vì input() đã trả về chuỗi
original_string = input("Nhập chuỗi bất kỳ: ") 

# Khởi tạo một chuỗi trống để lưu kết quả
result_string = "" 

# 2. Lặp qua từng ký tự trong chuỗi
for char in original_string:
    # Lấy mã ASCII của ký tự
    b = ord(char)
    
    # --- Logic chuyển đổi dựa trên mã ASCII ---
    
    # 2a. Nếu là chữ HOA (ASCII 65-90): chuyển sang chữ thường (+32)
    if 65 <= b <= 90:
        c = b + 32
        d = chr(c)
        
    # 2b. Nếu là chữ thường (ASCII 97-122): chuyển sang chữ HOA (-32)
    elif 97 <= b <= 122:
        c = b - 32
        d = chr(c)
        
    # 2c. Nếu không phải chữ cái: giữ nguyên ký tự đó (ví dụ: số, dấu cách, ký tự đặc biệt)
    else:
        d = char
        
    # Nối ký tự đã chuyển đổi vào chuỗi kết quả
    result_string += d 

# 3. In kết quả
print("Chuỗi sau khi đổi là:", result_string)
```

```python
#bài 8
GIA_DIEN = [
    (50, 1984),    # Bậc 1: 0 - 50 kWh
    (50, 2050),    # Bậc 2: 51 - 100 kWh (50 kWh tiếp theo)
    (100, 2380),   # Bậc 3: 101 - 200 kWh (100 kWh tiếp theo)
    (100, 2998),   # Bậc 4: 201 - 300 kWh (100 kWh tiếp theo)
    (100, 3350),   # Bậc 5: 301 - 400 kWh (100 kWh tiếp theo)
    (float('inf'), 3460) # Bậc 6: Từ 401 kWh trở đi
]
VAT_RATE = 0.08

# --- NHẬP DỮ LIỆU ---
try:
    ten = input("Nhập Họ và tên chủ hộ: ")
    cs_truoc = int(input("Nhập chỉ số điện kế tháng trước: "))
    cs_sau = int(input("Nhập chỉ số điện kế tháng này: "))
    
    so_dien_tt = cs_sau - cs_truoc
    if so_dien_tt < 0:
        print("Lỗi: Số điện tiêu thụ không hợp lệ.")
    else:
        tien_chua_vat = 0
        kwh_con_lai = so_dien_tt
        
        # 2. TÍNH TIỀN THEO BẬC THANG
        for kwh_bac, gia in GIA_DIEN:
            if kwh_con_lai <= 0:
                break
                
            # Số kWh sẽ được tính trong bậc hiện tại
            kwh_tinh = min(kwh_con_lai, kwh_bac)
            
            # Tính tiền bậc này và cộng dồn
            tien_chua_vat += kwh_tinh * gia
            
            # Trừ số kWh đã tính ra khỏi số kWh còn lại
            kwh_con_lai -= kwh_tinh
        
        # 3. TÍNH VAT VÀ LÀM TRÒN
        tong_tien_phai_tra = tien_chua_vat * (1 + VAT_RATE)
        
        # Chỉ lấy đến đơn vị đồng (làm tròn xuống)
        tien_lam_tron = int(tong_tien_phai_tra)
        
        # 4. IN KẾT QUẢ
        print(f"\n--- KẾT QUẢ ---")
        print(f"Họ và tên: {ten}")
        print(f"Tiền phải trả (đã bao gồm VAT): {tien_lam_tron:,} đồng")

except ValueError:
    print("Lỗi: Chỉ số điện kế phải là số nguyên.")
    
```
