```python
#W2A17
n = int(input("Nhập so nguyen n: "))

# Chiều rộng tối đa cố định (Max Width) của hàng cuối cùng
# Công thức: 2*n - 1
max_width = 2 * n - 1

# Lặp qua n hàng. i sẽ chạy từ 1 đến n (số thứ tự của hàng)
for i in range(1, n + 1):
    
    # 1. Tính số ký tự Z (luôn là số lẻ: 1, 3, 5, ...)
    num_chars = 2 * i - 1
    a = "Z" * num_chars
    
    print( a.center(max_width, "-"))
```
